## 17. Manipulação de Exceções - Tratamento de Erros

Exceção é um objeto que representa um erro que ocorreu ao executar o código/programa. Para manipular isso usaremos blocos do típo *try* ... *except*

#### 17.1 Exemplo com uma divisão por 0

```python
n1 = int(input('Digite um número: '))
n2 = int(input('Digite outro número: '))

try:
    r = round(n1 / n2, 2)
except ZeroDivisionError:
    print('Não é possível dividir por 0!\n')
else:
    print(f'Resultado: {r}\n')
```

#### 17.2 Exemplo com erro de valor

```python
def divisao(j, k):
    return round(j / k, 2)

if __name__ == '__main__':
    while True:
        try:
            n1 = int(input('Digite um número: '))
            n2 = int(input('Digite outro número: '))
            break
        except ValueError:
            print('Você não digitou um número. Tente novamente.')

    try:
        r = divisao(n1, n2) 
    except ZeroDivisionError:
        print('Não é possível dividir por zero!')
    except:
        print('Ocorreu um erro, reinicie o programa.')
    else:
        print(f'Resultado: {r}\n')
    finally: 
        print('Fim do cálculo')
```

#### 17.3 Exceções *raise*

```python
from math import sqrt

class numero_negativo_error(Exception):
    def __init__(self):
        pass

if __name__ == '__main__':
    while True:
        try:
            num = int(input('Digite um número positivo: '))
            if num < 0:
                raise numero_negativo_error
        except numero_negativo_error:
            print('\nVocê digitou um número negativo. Tente novamente.')
        except ValueError:
            print('Você não digitou um número. Tente novamente.\n')
        else:
            print(f'\nA raiz quadrada de {num} é {sqrt(num)}.\n')
        finally:
            print('Fim do cálculo\n')
```

## 18. Funções Recursivas
É uma técnica onde uma função chama a si mesma para resolver um problema dividindo-o em subproblemas menores e semelhantes.

#### 18.1 Exemplo calculando fatorial

A fórmula é definida por: 
$n! = n * (n-1)!$

```python
def fatorial(n):
    if n <= 1:
        return 1
    return n * fatorial(n - 1)

print(fatorial(5))  # 120 (5 * 4 * 3 * 2 * 1)
```

#### 18.2 Exemplo calculando a sequência de Fibonacci
A fórmula é definida por:
$F(n) = F(n-1) + F(n-2)$

```python
def fibonacci(n):
    if n < 0:
        return []
    elif n == 0:
        return [0]
    
    sequencia = [0, 1]
    for i in range(2, n + 1):
        sequencia.append(sequencia[-1] + sequencia[-2])
    return sequencia

print(fibonacci(10))
```

## 21. Programação Orientada a Objetos

POO é um paradigma de programação que organiza o código em torno de objetos que contêm dados (atributos) e comportamentos (métodos). O Python é uma linguagem *multiparadigma (estruturado, funcional, poo)*. 

#### 21.1 Classes

Classe é ... 

```python
class veiculo:
    def movimento(self):
        print(f'sou um carro...')
    
    def __init__(self, fabricante, modelo):
        self.fabricante = fabricante # Atributo público
        self.modelo = modelo # Atributo público
        self.num_registro = None # Atributo público

if __name__ == '__main__':
    meu_veiculo = veiculo('GM', 'Cruze')
    meu_veiculo.movimento()
    print(meu_veiculo.fabricante)
```

#### 21.1.1 Getter

Para criar atributos privados devemos usar a sintaxe: **self.__fabricante** (com 2 underscores na frente). E devemos usar o método *Getter* para acessar os atributos provados. 

```python
class veiculo:
    def movimento(self):
        print(f'sou um carro...')
    
    def __init__(self, fabricante, modelo):
        self.__fabricante = fabricante # Atributo privado
        self.__modelo = modelo # Atributo privado
        self.__num_registro = None # Atributo privado

    # Getter
    def get_fabricante_modelo(self):
        print(f'Fabricante: {self.__fabricante}, Modelo: {self.__modelo}\n')

if __name__ == '__main__':
    meu_veiculo = veiculo('GM', 'Cruze')
    meu_veiculo.movimento()
    meu_veiculo.get_fabricante_modelo()
```

#### 21.1.2 Setter

No caso do atributo **self.__num_registro**, para gravarmos informações nele, devemos usar o método *Setter*, que permite gravar nesse tipo de atributos.

```python
class veiculo:
    def movimento(self):
        print(f'sou um carro...')
    
    def __init__(self, fabricante, modelo):
        self.__fabricante = fabricante # Atributo privado
        self.__modelo = modelo # Atributo privado
        self.__num_registro = None # Atributo privado
    
    # Setter
    def set_num_registro(self, registro):
        self.__num_registro = registro

    # Getter
    def get_fabricante_modelo(self):
        print(f'Fabricante: {self.__fabricante}, Modelo: {self.__modelo}\n')

    def get_num_registro(self):
        return self.__num_registro

if __name__ == '__main__':
    meu_veiculo = veiculo('GM', 'Cruze')
    meu_veiculo.movimento()
    meu_veiculo.get_fabricante_modelo()
    meu_veiculo.set_num_registro('1654-6') # se for número pode escrever direto sem as aspas
    print(f'Registro: {meu_veiculo.get_num_registro()}\n')
```

#### 21.2 Herança

#### 22.1 Problema de exemplo 
Problema: Precisamos renomear todos os arquivos de uma determinada pasta seguindo um padrão informado pelo usuário.

```python
import os

diretorio = input('Digite o diretório dos arquivos: ')

os.chdir(diretorio)
print(f'Dir: {os.getcwd()}')

padrao_nome = input('Digite o novo padrão de nomes: ')

for contador, arquivo in enumerate(os.listdir()):
    if os.path.isfile(arquivo):
        nome_arquivo, ext_arquivo = os.path.splitext(arquivo)
        nome_arquivo = padrao_nome + '_' + str(contador)

        nome_novo = f'{nome_arquivo}{ext_arquivo}'
        os.rename(arquivo, nome_novo)

print(f'Arquivos renomeados com sucesso!')
```

Obs.: ainda poderiamos selecionar apenas arquivos com determinadas extensões, ou modificar o número inicial do contador, ou ainda copiar os novos arquivos para outro diretório mantendo os originais sem alterações, dentre várias outras possíbilidades. 

## 23. Manipulação de Arquivos de Texto - Leitura e Escrita

| Modo | Significado |
|----------|---------------|
| r | Somente leitura |
| r+ | Leitura e escrita. O texto é inserido no **início** do arquivo |
| w | Escrita, apagando (truncando) o conteúdo existente no arquivo |
| w+ | Leitura e escrita. O arquivo é criado, se não existir; se existi, é truncado. O texo é inserido no **início** do arquivo |
| x | Abre o arquivo para criação exclusiva, falhando se o arquivo já existir |
| a | Escrita, preservando o conteúdo existente (append). O arquivo é criado, se não existir. O texto é inserido no **final** do arquivo |
| b | Modo binário |
| t | modo texto (padrão) |
| + | Atualização - leitura e escrita |

- Modo padrão é 'r' (abre para leitura de texto, um sinônimo de 'rt'). 
- Modos 'w+' e 'w+b' abrem e truncam o arquivo. 
- Modos 'r+' e 'r+b' abrem o arquivo sem truncar o mesmo.

#### 24.5 Gerenciamento de contexto com with( )
with expressão as variável:

    # bloco de código

```python
# Sem with (manual)
file = open('arquivo.txt', 'r')
try:
    conteudo = file.read()
finally:
    file.close()  # Precisa fechar manualmente

# Com with (automático)
with open('arquivo.txt', 'r') as arquivo:
    conteudo = arquivo.read()
# Arquivo é fechado automaticamente ao sair do bloco
```

#### 24.5.1 Usando classe com `__enter__` e `__exit__`

```python
class GerenciadorArquivo:
    def __init__(self, nome_arquivo, modo):
        self.nome_arquivo = nome_arquivo
        self.modo = modo
        self.arquivo = None
    
    def __enter__(self):
        print(f"Abrindo arquivo {self.nome_arquivo}")
        self.arquivo = open(self.nome_arquivo, self.modo)
        return self.arquivo
    
    def __exit__(self, tipo_excecao, valor_excecao, traceback):
        print(f"Fechando arquivo {self.nome_arquivo}")
        if self.arquivo:
            self.arquivo.close()
        # Retornar True suprime exceções
        return False

with GerenciadorArquivo('teste.txt', 'w') as arquivo:
    arquivo.write('Olá mundo!')
```
