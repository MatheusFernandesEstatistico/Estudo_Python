# Quanto custa para completar o Álbum da Copa 2026 ?

```python
import random

valor_pacote = 7 
valor_album = [24.9, 74.9]
qnt_pacotes = 7
album_completo = 980

def completa_albuns(album_completo = 980, qnt_pacotes = 7):
    pct_comprados = 0
    album = set()

    while len(album) < album_completo:
        pacote = {random.randint(1, album_completo) for i in range(7)}
        album = album.union(pacote)
        pct_comprados += 1
    return pct_comprados
```

| Item  | Valor          | Quantidade / item |
|----------|---------------|---------|
| Pacote        | R$ 7.0        | 7 Figurinhas   |
| Álbum comum       | R$ 24.9     | 980 figurinhas   |
| Álbum capa dura       | R$ 74.9     | 980 figurinhas   |


#### 1. Completar sozinho

```python
albuns = [completa_albuns() for i in range(10000)]

import matplotlib.pyplot as plt
import seaborn as sns 

plt.hist(albuns, bins=20, 
         facecolor='#00aca0',
         edgecolor='black')
plt.title('Histograma de Albuns Completos')
plt.xlabel('Quantidade de Pacotes')
plt.ylabel('Quantidade de Albuns')
plt.show()
```

```python
import pandas as pd

mediana = pd.Series(albuns).median()
gasto_pacotes = mediana * valor_pacote
gasto_geral = gasto_pacotes + valor_album[0]

print(f'O total gasto com pacotes é R${gasto_pacotes}.')
```
